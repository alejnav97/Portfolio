```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\aleja\OneDrive\Documents/Movies.csv')
```

```python
# Taking A Peak At The Data 

df.head()
```

```python
# Checking For Missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
# Filling In MIssing Data 

df = df.fillna(0)
```

```python
# Column Data Type

df.dtypes
```

```python
# Converting Data Type of Colomns 

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
                                 
```

```python
df
```

```python
# Creating Column For Correct Year 

df['yearcorrect'] = df['released'].astype(str).str.split(',').str[1].str.split('(').str[0]
```

```python
df
```

```python
# Getting A Look At ALl Of The Data 

pd.set_option('display.max_rows', None)
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Comparing Budget Vs Gross using Scatter Plot 

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.ylabel('Gross Earnings')

plt.xlabel('Budget for Film')

df.fillna(0, inplace=True)

# Applying conditional Function To The Column Containing Negative Numbers

df['gross'] = df['gross'].apply(lambda x: abs(x) if x < 0 else x)

plt.show()
```

```python
df.head()
```

```python
# Using Seaborn To Plot Budget VS Gross To Find Correlation

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "black"})
```

```python
# Correlations Within the Data

df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.ylabel('Movie Feature')

plt.xlabel('Movie Feature')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.ylabel('Movie Feature')

plt.xlabel('Movie Feature')

plt.show()
```
